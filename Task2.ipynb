```python
df.fillna({'LAST_KNOWN_DELVRY_TYPE_CD':0}, inplace=True)
df.fillna({'ENGINE_SOURCE_PLANT':'NOT_SPECIFIED'}, inplace=True)
df.fillna({'ENGINE_TRACE_NBR':'NOT_SPECIFIED'}, inplace=True)
df.fillna({'TRANSMISSION_TRACE_NBR':'NOT_SPECIFIED'}, inplace=True)
df.fillna({'TRANSMISSION_SOURCE_PLANT':0}, inplace=True)


df.isna().sum()
```

```python
Q1 = df['TOTALCOST'].quantile(0.25)
Q3 = df['TOTALCOST'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['TOTALCOST'] >= lower_bound) & (df['TOTALCOST'] <= upper_bound)]
```

```python
!pip install openpyxl --upgrade
def safe_str(obj):
    """Handles objects that might cause recursion errors during string conversion."""
    try:
        return str(obj)
    except RecursionError:
        return "RecursionError_Object"  # or any other suitable placeholder

# Assuming 'problematic_column' is the name of the problematic column(s)
# Adjust this if you have multiple columns causing issues
for column in df.select_dtypes(include=['object']).columns:  # Loop through object-type columns
    df[column] = df[column].apply(safe_str)

# Now proceed with exporting to Excel
df.to_excel('cleaned_datan.xlsx', index=False, engine='openpyxl')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
top_repairs = df['GLOBAL_LABOR_CODE_DESCRIPTION'].value_counts().head(10)
print("Most common repairs:")
print(top_repairs)

top_repairs.plot(kind='bar', title='Top 10 Repair Types')
plt.show()
```

```python
cost_by_platform = df.groupby('PLATFORM')['TOTALCOST'].agg(['mean','median','count'])
print("\nRepair costs by platform:")
print(cost_by_platform.sort_values('mean', ascending=False))

df.boxplot(column='TOTALCOST', by='PLATFORM', rot=90)
plt.title('Repair Cost Distribution by Platform')
plt.show()
```

```python
state_cost_summary = df.groupby('STATE')['TOTALCOST'].agg(['sum'])


state_cost_summary = state_cost_summary.sort_values('sum', ascending=False)

print("State-wise Repair Cost Summary:")
print(state_cost_summary)
```

```python

state_cost_summary = state_cost_summary.reset_index()


plt.figure(figsize=(12, 6))
sns.barplot(x='STATE', y='sum', data=state_cost_summary, palette='viridis')

plt.title('Total Repair Cost by State', fontsize=16)
plt.xlabel('State', fontsize=12)
plt.ylabel('Total Repair Cost', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
